```python
sales_data_files = {
    'ventas_zip_92252': r'C:\Users\mcana\OneDrive\Escritorio\Power BI\AIRBNB MARKET & REAL ESTATE SALES DATA\datos\sales_properties_total_zipcode_92252.csv',
    'ventas_zip_92284': r'C:\Users\mcana\OneDrive\Escritorio\Power BI\AIRBNB MARKET & REAL ESTATE SALES DATA\datos\sales_properties_total_zipcode_92284.csv',
    'ventas_zip_92314': r'C:\Users\mcana\OneDrive\Escritorio\Power BI\AIRBNB MARKET & REAL ESTATE SALES DATA\datos\sales_properties_total_zipcode_92314.csv',
    'ventas_zip_92315': r'C:\Users\mcana\OneDrive\Escritorio\Power BI\AIRBNB MARKET & REAL ESTATE SALES DATA\datos\sales_properties_total_zipcode_92315.csv'
}
```

```python
sales_data = {}
price_stats = {}

for name, path in sales_data_files.items():
    df = pd.read_csv(path, delimiter=';')
    sales_data[name]= df
    
    # extraer el precio de información
    prices = df['Price']
    price_stats[name] = {
    'Mean Price': prices.mean(),
    'Median Price': prices.median(),
    'Min Price': prices.min(),
    'Max Price': prices.max()
}
```

```python
price_stats_df = pd.DataFrame(price_stats).transpose()

price_stats_df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,10))
for ax, (name, df) in zip(axs.flatten(), sales_data.items()):
    prices = df['Price']
    ax.hist(prices, bins=20, color='purple', alpha = 0.7)
    ax.set_title(f'price distribution for {name.split('_')[1]}')
    ax.set_xlabel('Price ($)')
    ax.set_ylabel('Frequency')
    
plt.tight_layout()
plt.show()
```

##### distribución de los precios:
- ZIP 92252 : la mayoria de los precios se encuentran en el rango más bajo
- ZIP 92284 : una distribución más uniforme con algunos precios más altos.
- ZIP 92314 : Muestra una tendencia hacia precios más altos.
- ZIP 92315 : Similar al 92314, con un rango medio más alto.

#### ANÁLISIS DE CORRELACIÓN ENTRE EL PRECIO Y OTRAS VARIABLES.

```python
correlation_results = {}
variables = ['Bathrooms', 'Bedrooms', 'Living Area','Lot Size']

for name, df in sales_data.items():
    df['Price'] = df['Price'].replace({r'\$':'', ',': ''}, regex=True).astype(float)
    
    for var in variables:
        df[var] = pd.to_numeric(df[var], errors='coerce')
        
    correlation = df[['Price'] + variables].corr()
    correlation_results[name] = correlation['Price'][1:]
    
correlation_df = pd.DataFrame(correlation_results).transpose()
correlation_df

    
```

##### Conclusión
- el área habitable muestra una fuerte correlación positiva con los precios, sobre todo 92314, 92252
- el nº de baños la correlación es mas variable, pero en 92314 es bastante alta
